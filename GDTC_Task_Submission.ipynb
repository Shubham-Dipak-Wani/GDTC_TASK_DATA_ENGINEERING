# Task #6 

#### Business wants to check the Claim Amount for deviation for each transaction, they would like you to calculate as follow
 

```python
Combined_Data.TXN_DATE_TIME = pd.to_datetime(Combined_Data['TXN_DATE_TIME'])
```

```python
Combined_Data.info()
```

```python
Combined_Data.TXN_DATE_TIME.max()
```

```python
Last_30days_Data = Combined_Data.loc[(Combined_Data['TXN_DATE_TIME'] > '2021-05-31') & (Combined_Data['TXN_DATE_TIME'] < '2021-06-30')]
```

```python
Last_30days_Data.head()
```

```python
#CLAIM_DEVIATION = AVG_CLAIM_AMOUNT_FOR_LAST_30DAYS (same insurance type)/ CURRENT_CLAIM_AMOUNT
Combined_Data['CLAIM_DEVIATION'] = Last_30days_Data['CLAIM_AMOUNT'].mean() / Combined_Data['CLAIM_AMOUNT']
Combined_Data.head()
```

```python
#If the value < 0.5 THEN CLAIM_DEVIATION = 1 ELSE 0 .If there is less than 30 days of transaction history THEN-1 .Note: LAST_30DAYS does not include current day
Combined_Data['CLAIM_DEVIATION']=np.where(Combined_Data['CLAIM_DEVIATION'] < 0.5,'1','0')
Combined_Data.head(10)
```

```python
Combined_Data.CLAIM_DEVIATION[Combined_Data.TXN_DATE_TIME.max() - Combined_Data.TXN_DATE_TIME < pd.Timedelta(days=30)] = -1 
```

```python
Combined_Data.CLAIM_DEVIATION.value_counts()
```

#### Business wants to check Claim Velocity which is defined as follow

```python
Claim_Data_last30days = Combined_Data.loc[(Combined_Data['TXN_DATE_TIME'] > '2021-05-31') & (Combined_Data['TXN_DATE_TIME'] < '2021-06-30')]
Claim_Data_last3days = Combined_Data.loc[(Combined_Data['TXN_DATE_TIME'] > '2021-06-27') & (Combined_Data['TXN_DATE_TIME'] < '2021-06-30')]
```

```python
Data_Combined.rename( columns = {"CLAIM_STATUS_x":"CLAIM_DATA_30_DAYS"}, inplace = True)
Data_Combined.rename( columns = {"CLAIM_STATUS_y":"CLAIM_DATA_3_DAYS"}, inplace =True)
Data_Combined
```

```python
Data_Combined['CLAIM_VELOCITY'] = Data_Combined['CLAIM_DATA_30_DAYS']/Data_Combined['CLAIM_DATA_3_DAYS']
Data_Combined
```
